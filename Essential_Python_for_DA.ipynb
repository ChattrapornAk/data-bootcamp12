### **OOP**

```python
class Dog:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age

    def bark(self):
        print(f"{self.name} is barking!")

    def hdb(self):
        self.age += 1
```

```python
# attributes
dog1 = Dog("Buddy", "Golden Retriever", 2)
dog2 = Dog("Max", "German Shepherd", 3)
```

```python
print(f"{dog1.name} is a {dog1.species}")
print(f"{dog2.name} is a {dog2.species}")
```

```python
dog1.bark()
```

```python
print(dog1.age)
dog1.hdb()
print(dog1.age)
```

```python
# Homework Sample
class ATM:
    def __init__(self, username, bank, amount):
        self.username = username
        self.bank = bank
        self.amount = amount

    def __str__(self):
        return f"{self.username} uses {self.bank} and has {self.amount} THB in an account."

    def check_balance(self):
        print(f"Balance: {self.amount} THB")

    def deposit(self, saving):
        self.amount += saving
        print(f"Deposit {saving} THB successfully.")
```

```python
somo = ATM("chattraporn", "scb", 100)
```

```python
print(f"{somo.username} opens a new bank account at {somo.bank}")
```

```python
somo.check_balance()
```

```python
somo.deposit(500)
```

```python
print(somo)
```

### **Super Class**

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

## extension : build on top Person
# super refer to Person
# Emplypee is super class build on top class
class Employee(Person):
    def __init__(self, name, age, company):
        super().__init__(name, age)
        self.company = company

    def greeting(self):
        print("Hello World")
```

```python
Person1 = Person("somo", 37)
```

```python
Person2 = Employee("Ahyeon", 18, "YG")
```

```python
Person2.greeting()
```

```python
# context manager
read_data = []

with open("hotel.csv", "r") as file:
    data = csv.reader(file)
    for row in data:
        read_data.append(row)
```

```python
#header
read_data[0]
```

```python
# body data
read_data[1: ]
```

```python
# use this syntax to calculate im original python
# for loop to calculate avg. price per night
price = []

for row in read_data[1: ]:
    price.append(int(row[4])) ## or use [-1] the last item -> negative index

print(f"Average price per night: {sum(price)/ len(price)}")
```

###**Request API**
How to get data via API

```python
import requests

#api endpoint
url = "https://swapi.info/api/people/1"
```

```python
# get requests
response = requests.get(url)
response
```

```python
# format to json form
response.json()["name"]
```

```python
# for loop
import requests
import time

names = []

for i in range(5):
    url = f"https://swapi.info/api/people/{i+1}"
    response = requests.get(url)
    response_json = response.json()
    names.append(response_json["name"])
    print(f"Success: {i+1}")
    time.sleep(1)

print(names)
```
