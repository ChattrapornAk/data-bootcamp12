### **map / filter**

```python
# iterable (adj)
scores = [82, 78, 81, 66, 50]
```

```python
# create funtion
def grading(score):
    if score >= 80:
        return "passed"
    else:
        return "failed"
```

```python
# use map with grading function
list(map(grading, scores))
```

```python
# add 5 to scores
list(map(lambda x: x+5, scores))
```

```python
# filter
scores
```

```python
# use for loop
students_score_over80 = []
for score in scores:
    if score > 80:
        students_score_over80.append(score)
print(students_score_over80)
```

```python
# use filter lambda
list(filter(lambda x: x>80, scores))
```

```python
scores = [ ("toy", 80), ("jane", 78), ("lisa", 95)]

# filter score>=80, [index]
list(filter(lambda x: x[1]>=80, scores))
```

## **Numpy**
Numerical python = matrix / computing

```python
import random

nums = []

for i in range(10):
    nums.append(random.randint(100,999))

print(nums)
```

```python
from typing_extensions import SupportsComplex
# naive python no averge, mean function
# can sum
sum(nums)
```

```python
# average in naive python cannot use mean()
sum(nums)/ len(nums)
```

```python
# so numpy can fix this problem in naive python
# numpy create for computing
import numpy as np

nums
```

```python
# numpy function
print(np.sum(nums),
      np.mean(nums),
      np.std(nums),
      np.median(nums),
      np.var(nums))
```

```python
# change to numpy array
# array == vector in R
nums = np.array(nums)
nums
```

```python
# after change numpy to array can use numpy.method
print(nums.sum(),
      nums.mean(),
      nums.min(),
      nums.max())
```

```python
# import library(module)
import requests
from gazpacho import Soup
```

```python
url = "https://raw.githubusercontent.com/toyeiei/python-test-bc12/refs/heads/main/index.html"
```

```python
# get request
response = requests.get(url)
```

```python
response.status_code
```

```python
# web scraping : load data html
web = Soup(response.text)
web
```

```python
# use web scraping to select data in html
# Soup method
web.find("h1")
```

```python
# delete all html tag : extract only text
for h2 in web.find("h2"):
    print(h2.strip())
```

```python
web.find("li")
```

```python
for li in web.find("li"):
    print(li.strip())
```
