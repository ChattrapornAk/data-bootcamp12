```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
df_california_texas_2017 = df[(df['Order Date'].between('2017-01-01', '2017-12-31')) & (df['State'].isin(['California', 'Texas']))]
df_california_texas_2017.to_csv('df_california_texas_2017.csv')
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
# Answer: total sales(484,247.49), average sales(242.97), std(623.21)
import numpy as np
df_2017 = df[df['Order Date'].between('2017-01-01', '2017-12-31')]
print(np.sum(df_2017['Sales']),
      np.mean(df_2017['Sales']),
      np.std(df['Sales']))
```

```python
# TODO 06 - which Segment has the highest profit in 2018
# Answer: Comsumer segment profit 3177.475
df_2018 = df[df['Order Date'].between('2018-01-01', '2018-12-31')]
df_2018.sort_values('Profit', ascending=False).head()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
# Answer: Top 5 states are New Hamshire, New maxico, Dist. of Columbia, Louisiana, South Coralina
df_150419_311219 = df[df['Order Date'].between('2019-04-15', '2019-12-31')]
df_150419_311219.groupby('State')['Sales'].sum().sort_values().head()
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
# Answer: 54.97%
df_2019 = df[df['Order Date'].between('2019-01-01', '2019-12-31')]
df2019_region = df_2019.groupby('Region')['Sales'].sum()

total_sales = df2019_region.sum()
proportion = (df2019_region['West'] + df2019_region['Central']) / total_sales
print(proportion *100)
```

```python
# TODO 09.1 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
# term of number of  orders
df_2019_2020 = df[df['Order Date'].between('2019-01-01', '2020-12-31')]
group_product = df_2019_2020.groupby('Product Name').agg(
    product_count = ('Product Name', 'count'),
    total_sales = ('Sales', 'sum')
)
group_product.sort_values('product_count', ascending=False).head(10)
```

```python
# TODO 09.2 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
# term of total sales
group_product.sort_values('total_sales', ascending=False).head(10)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# Answer: bar chart between total sales vs region
region_sales = df.groupby('Region').agg(total = ('Sales', sum))
region_sales['total']\
    .sort_values(ascending=False)\
    .plot(kind='bar', color=['#ffa69e', '#faf3dd', '#b8f2e6', '#aed9e0']);

# Answer: scatter plot between quantity vs sales
df[['Quantity', 'Sales']]\
    .plot(x='Quantity',y='Sales',kind='scatter', color='#ffa69e');
```
